Nombre de Columnas en snake_case

```python
df.columns= df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

Comprobar y tratar nulos

```python
df.isna().sum()
```

Mantener filas que tengan mas de 19 valores no nulos

```python
df = df.dropna(thresh=19)
```

```python
df.shape
```

Poner "0" a todos los campos nulos

```python
df = df.fillna(0)
```

```python
df.isna().sum()
```

Eliminar Duplicados

```python
df = df.drop_duplicates()
```

```python
df.shape
```

Limpiar Textos y normalizar valores

```python
df["neighbourhood"] = df["neighbourhood"].str.strip().str.lower()
```

```python
df["room_type"] = df["room_type"].str.strip().str.lower()
```

```python
df["neighbourhood"]
```

```python
df["room_type"]
```

Limpiar Precios

```python
df["price"] = df["price"].replace('[\$,]', '', regex = True).astype(float)
```

```python
df["price"]
```
