```python
import pandas as pd
```

```python
import numpy as np
```

IMPORTAR DATOS DE CSV

```python
df = pd.read_csv("personas_sucio.csv")
```

```python
df
```

NO HAY VALORES NULOS

TODOS LOS NOMBRES INICIAN POR MAYUSCULA Y EL RESTO EN MINUSCULA

```python
df["nombre"] = df["nombre"].str.title()
```

```python
df
```

```python
mapa_edades = {"veintiocho": 28, "treinta": 30}
```

```python
df["edad"] = df["edad"].replace(mapa_edades)
```

```python
df
```

```python
df["edad"] = pd.to_numeric(df["edad"], errors="coerce")
```

```python
df
```

```python
media_edad = round(df["edad"].mean())
```

```python
df["edad"] = df["edad"].fillna(media_edad)
```

```python
df
```

TODAS LAS CIUDADES ESCRITAS EN MAYUSCULAS

```python
df["ciudad"] = df["ciudad"].str.upper()
```

```python
df
```

TODOS LOS EMAILS CON FORMATO CORRECTO

```python
df["email"] = df["email"].fillna("no-emeil@example.com")
```

```python
df
```

eliminar espacios entre nombre y @

```python
df["email"] = df["email"].str.replace(r'(@)(\s+)', r'\1', regex=True)
```

```python
df
```

si el @ se repite

```python
df["email"] = df["email"].str.replace(r'(@+)', r'@', regex=True)
```

```python
df
```

añadir @ si no lo tiene

```python
df["email"] = df["email"].str.replace(r'(?<!@)(mail)', r'@mail', regex = True)
```

```python
df["email"] = df["email"].str.replace(r'(?<!@)(example)', r'@example', regex = True)
```

```python
df
```

añade .com al final de todos los correos

```python
df["email"] = df["email"].str.replace(r'(@mail)$', r'@mail.com', regex = True)
```

```python
df["email"] = df["email"].str.replace(r'(@example)$', r'@example.com', regex = True)
```

```python
df
```

añade el "." de .com

```python
df["email"] = df["email"].str.replace(r'(mail)(com)', r'\1.\2', regex=True)
```

```python
df
```

PONER FORMATO DE FECHA EN FECHA_REGISTRO

```python
df["fecha_registro"] = pd.to_datetime(df.get("fecha_registro"), errors="coerce")
```

```python
df
```

```python
df["fecha_registro"] = df["fecha_registro"].fillna("01/01/2000")
```

```python
df
```

```python
df["fecha_registro"] = pd.to_datetime(df["fecha_registro"], format='%d-%m-%Y', errors="coerce")
```

```python
df
```

```python
df["fecha_registro"] = df["fecha_registro"].dt.strftime('%d-%m-%Y')
```

```python
df
```

```python
df.to_csv("personas_limpio.csv", index=False)
```
